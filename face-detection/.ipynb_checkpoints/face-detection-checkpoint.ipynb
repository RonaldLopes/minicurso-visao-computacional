## O algoritmo Haar Cascade

O algoritmo Haar Cascade  uma tcnica de deteco de objetos amplamente utilizada em viso computacional para identificar objetos especficos em imagens ou fluxos de vdeo. Ele  notvel por sua eficincia e capacidade de deteco rpida, tornando-o adequado para aplicaes em tempo real. O nome "Haar Cascade"  derivado dos filtros Haar usados na tcnica.

 

A deteco de objetos por Haar Cascade  frequentemente usada em aplicaes como reconhecimento facial, deteco de veculos, identificao de objetos em imagens mdicas, entre outros. Aqui esto alguns pontos-chave sobre como o algoritmo funciona:

- **Classificadores em cascata**: A deteco de objetos Haar Cascade envolve o uso de classificadores em cascata. Um classificador em cascata  composto por vrias etapas, cada uma das quais  responsvel por uma tarefa especfica na deteco do objeto. Cada etapa, ou estgio,  um classificador fraco que determina se uma parte da imagem corresponde ao objeto de interesse.

- **Caractersticas de Haar**: As caractersticas de Haar so padres retangulares que so usados para descrever as propriedades locais da imagem. Essas caractersticas so calculadas em reas retangulares adjacentes na imagem e so usadas para distinguir entre objetos e no objetos. As caractersticas de Haar podem representar bordas, linhas, texturas e outras informaes visuais.

- **Treinamento do classificador**: O processo de criao de um classificador Haar Cascade envolve o treinamento com um grande conjunto de imagens positivas (imagens que contm o objeto-alvo) e negativas (imagens que no contm o objeto-alvo). O classificador  treinado para aprender as caractersticas distintivas do objeto a ser detectado.

- **Cascata de estgios**: A ideia-chave por trs da eficincia do Haar Cascade  que a maioria das regies da imagem no contm o objeto-alvo. Portanto, o algoritmo utiliza uma cascata de estgios, onde a maioria das regies da imagem  rapidamente descartada nas primeiras etapas. Apenas as regies que passam por todas as etapas so consideradas positivas e indicativas da presena do objeto.

- **Ajuste de parmetros**: O desempenho do classificador pode ser ajustado por meio de parmetros, como o nmero de estgios na cascata, o tamanho das caractersticas de Haar e o limiar de deciso. A otimizao desses parmetros  importante para equilibrar a sensibilidade e a especificidade da deteco.

## Bibliotecas
Para realizar este estudo, ser necessrio importar algumas bibliotecas essenciais que desempenharo papis cruciais, como carregar a imagem, realizar pr-processamento e efetuar a deteco do objeto-alvo.

```python
from PIL import Image
import cv2
import numpy as np
import requests  
import matplotlib.pyplot as plt
```

## Carregando e processando a imagem
Em seguida, precisamos ler a imagem usando a funo imread() do OpenCV:

```python
imagePath = 'images/pessoas.png'
img = cv2.imread(imagePath)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

Isso carregar a imagem do caminho especificado e a retornar na forma de uma matriz Numpy.

Vamos imprimir as dimenses desta matriz:

```python
img.shape
```

O resultado ser algo parecido com (600, 800, 3), onde 600  a altura, 800  a largura e 3 so os canais de cores (BGR). Observe que o OpenCV usa o layout BGR (Azul, Verde, Vermelho) em vez do layout convencional RGB (Vermelho, Verde, Azul).

Para melhorar a eficincia computacional, primeiro precisamos converter esta imagem para tons de cinza antes de realizar a deteco de rosto:

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape
```

O resultado ser algo como (600, 800), pois a imagem agora  em tons de cinza e no possui mais o terceiro canal de cores.

```python
# from google.colab.patches import cv2_imshow
# cv2_imshow(imagem)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.show()
```

## Carregando o classificador
Vamos carregar o classificador Haar Cascade pr-treinado que est includo no OpenCV:

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

Observe que estamos usando um arquivo chamado "haarcascade_frontalface_default.xml". Este classificador  projetado especificamente para detectar rostos frontais em entradas visuais.

O OpenCV tambm fornece outros modelos pr-treinados para detectar objetos diferentes em uma imagem, como olhos, sorrisos, parte superior do corpo de uma pessoa e at mesmo a placa de um veculo. Voc pode aprender mais sobre os diferentes classificadores includos no OpenCV examinando o repositrio do OpenCV.

## Realizando a deteco
Agora podemos realizar a deteco de rosto na imagem em tons de cinza usando o classificador que acabamos de carregar:

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
```

Essa funo detectMultiScale()  usada para identificar rostos de tamanhos diferentes na imagem de entrada.

- **gray_image**: O primeiro parmetro  a imagem em tons de cinza que criamos anteriormente.

- **scaleFactor**: Este parmetro  usado para redimensionar o tamanho da imagem de entrada, tornando mais fcil para o algoritmo detectar rostos maiores. Neste caso, especificamos um fator de escala de 1.1, indicando que queremos reduzir o tamanho da imagem em 10%.

- **minNeighbors**: O classificador em cascata aplica uma janela deslizante na imagem para detectar rostos nela. Inicialmente, o classificador pode capturar um grande nmero de falsos positivos. Esses falsos positivos so eliminados usando o parmetro minNeighbors, que especifica o nmero de retngulos vizinhos que precisam ser identificados para que um objeto seja considerado uma deteco vlida.

- **minSize**: Por fim, o parmetro minSize define o tamanho mnimo do objeto a ser detectado. O modelo ignorar rostos menores do que o tamanho mnimo especificado.


Agora que o modelo detectou os rostos na imagem, vamos criar uma caixa delimitadora ao redor desses rostos:

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

A varivel face  uma matriz com quatro valores: as coordenadas x e y onde os rostos foram detectados e suas larguras e alturas. O cdigo acima itera sobre os rostos identificados e cria uma caixa delimitadora que abrange essas medidas. O parmetro (0, 255, 0) representa a cor da caixa delimitadora, que  verde, e 4 indica sua espessura.

Para exibir a imagem com os rostos detectados, primeiro precisamos converter a imagem do formato BGR para RGB:

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB))
plt.show()
```
